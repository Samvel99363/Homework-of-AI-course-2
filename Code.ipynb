```python
graph = {
    'Yerevan': {'Gyumri': 20, 'Sevan': 60},
    'Gyumri': {'Yerevan': 20, 'Vanadzor': 40},
    'Sevan': {'Yerevan': 60, 'Dilizhan': 40},
    'Vanadzor': {'Gyumri': 40, 'Dilizhan': 50},
    'Dilizhan': {'Sevan': 40, 'Vanadzor': 50}
}

heuristics = {
    'Yerevan': 90,
    'Gyumri': 85,
    'Sevan': 25,
    'Vanadzor': 35,
    'Dilizhan': 0
}
```

```python
import heapq

def a_star_search(graph, start, goal, heuristics):
    # Priority queue: (f_value, g_value, path)
    queue = [(heuristics[start], 0, [start])]
    visited = set()

    while queue:
        f, g, path = heapq.heappop(queue)
        current = path[-1]

        if current == goal:
            return path, g

        if current in visited:
            continue
        visited.add(current)

        for neighbor, cost in graph[current].items():
            new_g = g + cost
            h = heuristics[neighbor]
            f_value = new_g + h
            new_path = path + [neighbor]
            heapq.heappush(queue, (f_value, new_g, new_path))

    return None, float('inf')
```

```python
path, cost = a_star_search(graph, 'Yerevan', 'Dilizhan', heuristics)
print("Path found by A*:", path)
print("Total cost:", cost)
```
